```python
import h5py
import numpy as np
import pandas as pd
import json
```

```python
files = [
    "val_18k.h5",
    "val_700k_strong.h5",
    "val_9k.h5",
    "val_cnn_lstm_sa_mlp.h5",
    "val_cnn_lstm_sa.h5",
    "val_cnn_lstm.h5",
    "val_lstm.h5"
]
```

```python
scene_file = 'CLEVR_val_scenes.json'
with open(scene_file) as f:
    scenes = json.load(f)
    scenes_info = scenes["info"]
    scenes = scenes["scenes"]
question_file = 'CLEVR_val_questions.json'
with open(question_file) as f:
    questions = json.load(f)
    questions_info = questions["info"]
    questions = questions["questions"]
print("Questions info: ", questions_info)
print("Scenes_info: ", scenes_info)
```

```python
df = pd.DataFrame([])
```

```python
for file in files:
    f = h5py.File(file, "r")
    keys = f.keys()
    print("keys of ", file, ": ", keys)
    column_name = file[4:-3]
    df[column_name + "_correct"] = f["correct"][:].flatten()
    df[column_name + "_preds"] = f["preds"][:].flatten()
```

```python
print(questions[0].keys())
print(questions[10]["answer"])
```

```python
df["image_index"] = ""
df["question_index"] = ""
df["question_family_index"] = ""
df["answer"] = ""
df["question"] = ""
df["image_filename"] = ""
for i in range(df.shape[0]):
    df.at[i, "image_index"] = questions[i]["image_index"]
    df.at[i, "question_index"] = questions[i]["question_index"]
    df.at[i, "question_family_index"] = questions[i]["question_family_index"]
    df.at[i, "answer"] = questions[i]["answer"]
    df.at[i, "question"] = questions[i]["question"]
    df.at[i, "image_filename"] = questions[i]["image_filename"]
```
